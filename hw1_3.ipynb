LOWERCASE

```python
for i in range (len(data)):
    words=data.loc[i,'text']
    words=words.lower()
    data.loc[i,'text']=words
```

REMOVE PUNCTUATIONS

```python
import string
```

```python
s=string.punctuation
```

```python
for i in range(len(data)):
    s1=""
    for ch in data.loc[i,'text']:
        if ch not in s:
            s1+=ch
    data.loc[i,'text']=s1
```

REMOVE STOPWORDS

```python
import nltk
```

```python
nltk.download('stopword')
```

```python
from nltk.corpus import stopwords
stop_words=set(stopwords.words('english'))
```

```python
for i in range(len(data)):
    word=[]
    s=data.loc[i,'text']
    words1=s.split(' ')
    for words in words1:
        if words not in stop_words:
            word.append(words)
    data.loc[i,'text']=' '.join(word)
```

LEMMETISATION

```python
nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer
lemmetizing=WordNetLemmatizer()
for i in range(len(data)):
    s =data.loc[i,'text']
    words=s.split(' ')
    ans=[]
    for word in words:
        ans.append(lemmetizing.lemmatize(word))
    data.loc[i,'text']=' '.join(ans)
```

TOKENIZATION

```python
from nltk.tokenize import word_tokenize

for i in range(len(data)):
    sen=data.loc[i,'text']
    data.loc[i,'text']==word_tokenize(sen)
```
