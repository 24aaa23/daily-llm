```python
import torch
import torch.nn as nn
```

```python
import pandas as pd
import numpy as np
```

```python
from  sklearn.preprocessing import  StandardScaler
from  sklearn.preprocessing import  LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv")
```

```python
data.head()
```

```python
data=data.drop(columns=['id','Unnamed: 32'])
```

```python
data.shape
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(data.iloc[:,1:],data.iloc[:,:1],test_size=0.2)
```

```python
xtrain.head()
```

```python
xtrain.shape
```

```python
scaler=StandardScaler()
encoder=LabelEncoder()
```

```python
xtrain =scaler.fit_transform(xtrain)
xtest=scaler.transform(xtest)
ytrain=encoder.fit_transform(ytrain)
ytest=encoder.transform(ytest)
```

```python
class Model(nn.Module):
    def __init__(self,newfeature):
        super().__init__()
        self.linear=nn.Linear(newfeature,16)
        self.relu=nn.ReLU()
        self.linear1=nn.Linear(16,1)
        self.sigmoid=nn.Sigmoid()
    def forward(self,feature):
        out=self.linear(feature)
        out=self.relu(out)
        out=self.linear1(out)
        out=self.sigmoid(out)
        return out
```

transformation from numpy to tensors

```python
xtrain=torch.from_numpy(xtrain.astype(np.float32))
xtest=torch.from_numpy(xtest.astype(np.float32))
ytrain=torch.from_numpy(ytrain.astype(np.float32))
ytest=torch.from_numpy(ytest.astype(np.float32))
```

```python
learning_rate=0.1
epoch=20
loss_function=nn.BCELoss()
```

```python
model=Model(xtrain.shape[1])
```

```python
optimiser=torch.optim.SGD(model.parameters(),learning_rate)
```

```python
for epoches in range(epoch):
    ypred=model(xtrain)#forwardpass
    #loss
    loss=loss_function(ypred,ytrain.view(-1,1))
    optimiser.zero_grad()
    loss.backward()
    optimiser.step()
    print(f'epoch:{epoches+1},Loss:{loss.item()}')

```
