```python
import torch
import torch.nn as nn
```

```python
import pandas as pd
import numpy as np
from sklearn .preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv")
data.head()
```

```python
data=data.drop(columns=['id','Unnamed: 32'])
```

```python
scalar=StandardScaler()
encoder=LabelEncoder()
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(data.iloc[:,1:],data.iloc[:,:1],test_size=0.2)
```

```python
xtrain.head()
```

```python
xtrain=scalar.fit_transform(xtrain)
xtest=scalar.transform(xtest)
ytrain=encoder.fit_transform(ytrain)
ytest=encoder.transform(ytest)
```

```python
xtrain
```

```python
xtrain1=torch.from_numpy(xtrain.astype(np.float32))
xtest1=torch.from_numpy(xtest.astype(np.float32))
ytrain1=torch.from_numpy(ytrain.astype(np.float32))
ytest1=torch.from_numpy(ytest.astype(np.float32))
```

```python
class Model(nn.Module):
    def __init__(self,features):
        super().__init__()
        self.fc1=nn.Linear(features,16)
        self.relu=nn.ReLU()
        self.fc2=nn.Linear(16,1)
        self.sigmoid=nn.Sigmoid()
    def forward(self,new_feature):
        out=self.fc1(new_feature)
        out=self.relu(out)
        out=self.fc2(out)
        out=self.sigmoid(out)
        return out

```

```python
t=xtrain1.shape[1]
t
```

```python

learning_rate=0.2
epoches=20
```

```python
loss=nn.BCELoss()
model=Model(t)
optimiser=torch.optim.SGD(model.parameters(),learning_rate)
```

```python
for i in range(epoches):
    #forward pass
    ypred=model(xtrain1)
    lossf=loss(ypred,ytrain1.view(-1,1))
    optimiser.zero_grad()
    lossf.backward()
    optimiser.step()
    print(f'the accuracy in {i+1} epoches is loss:{lossf.item()}' )

    
```
