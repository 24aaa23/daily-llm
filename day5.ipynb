```python
import torch
import torch.nn as nn
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv")
```

```python
data.head()
```

```python
data=data.drop(columns=['id','Unnamed: 32'])
```

```python
data.shape
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(data.iloc[:,1:],data.iloc[:,:1])
```

```python
scaler=StandardScaler()
encoder=LabelEncoder()
```

```python
xtrain=scaler.fit_transform(xtrain)
xtest=scaler.transform(xtest)
ytrain=encoder.fit_transform(ytrain)
ytest=encoder.transform(ytest)
```

```python
xtrain
```

```python
xtrain=torch.from_numpy(xtrain.astype(np.float32))

xtest=torch.from_numpy(xtest.astype(np.float32))
ytrain=torch.from_numpy(ytrain.astype(np.float32))
yest=torch.from_numpy(ytest.astype(np.float32))
```

```python
xtrain
```

```python
class Model(nn.Module):
    def __init__(self,newfeature):
        super().__init__()
        self.linear=nn.Linear(newfeature,16)
        self.relu=nn.ReLU()
        self.linear1=nn.Linear(16,1)
        self.sigmoid=nn.Sigmoid()
    def forward(self,feature):
        out=self.linear(feature)
        out=self.relu(out)
        out=self.linear1(out)
        out=self.sigmoid(out)
        return out

```

```python
t=xtrain.shape[1]

t
```

```python
learning_rate=0.1
epochs=30

loss=nn.BCELoss()
model=Model(t)
```

```python

optimizer=torch.optim.SGD(model.parameters(),learning_rate)
```

```python
for i in range(epochs):
    ypred = model(xtrain)
    loss_value = loss(ypred, ytrain.view(-1, 1))
    optimizer.zero_grad()
    loss_value.backward()
    optimizer.step()
    print(f'Loss at epoch {i}: {loss_value.item():.4f}')
```

```python
from sklearn.datasets import make_classification
```

```python
x,y= make_classification(
    n_samples=100,
    n_features=30,
    n_informative=2,
    n_redundant=0,
    n_classes=2,
    random_state=42,
)
```

```python
xtrain =torch.tensor(xtrain,dtype=torch.float32)
y=torch.tensor(y,dtype=torch.long)
from torch.utils.data import Dataset,DataLoader
```

```python
class CustomDataset(Dataset):
    def __init__(self,features,labels):
        self.features=features
        self.labels=labels
    def __len__(self):
        return self.features.shape[0]
    def __getitem__(self,index):
        return self.features[index],self.labels[index]
    
```

```python
dataset=CustomDataset(x,y)
```

```python
dataloader=DataLoader(dataset,batch_size=4,shuffle=True)
```
