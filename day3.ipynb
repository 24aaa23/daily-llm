```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv")
```

```python
data.head()
data=data.drop(columns=['id','Unnamed: 32'])
```

```python
xtrain ,xtest,ytrain,ytest= train_test_split(data.iloc[:,1:],data.iloc[:,:1] ,train_size=0.2)
```

```python
xtrain.head()
```

```python
scaler=StandardScaler()
encoder=LabelEncoder()
xtrain =scaler.fit_transform(xtrain)
xtest=scaler.transform(xtest)
ytrain=encoder.fit_transform(ytrain)
ytest=encoder.fit_transform(ytest)
```
